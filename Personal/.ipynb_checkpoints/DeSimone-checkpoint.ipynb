```python
import pandas as pd

# Load the CSV file
# Replace 'data.csv' with the path to your CSV file
df = pd.read_csv('../Datasets/Fire_Incidents.csv')

# Count occurrences of each neighborhood
neighborhood_counts = df['neighborhood'].value_counts()

# Display the counts for each neighborhood
print (neighborhood_counts)
```

```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ['Ridgemont', 'Regent Square', 'Mt.Oliver', 'Mount Oliver Borough', 'East Carnegie']
values = [12, 9, 12, 6, 11]

# Create the bar chart
plt.bar(categories, values, color='orange')

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right')  # Rotate 45 degrees, align text to the right

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Top 5 Lowest Neighborhood Numbers of Fires')

# Adjust layout to prevent clipping
plt.tight_layout()

# Display the graph
plt.show()
```
