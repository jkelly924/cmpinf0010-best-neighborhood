# Plotting Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# read the CSV file
df = pd.read_csv('Neighborhoods.csv')

# Select specific columns using iloc
filtered_df = df.iloc[:, [21, 23]]

# Rename columns 
filtered_df.columns = ['hood', 'acres']

# converting acres to a numeric
filtered_df['acres'] = pd.to_numeric(filtered_df['acres'], errors='coerce')



# Sort by 'acres' in ascending order and select top 10
top_10 = filtered_df.sort_values(by='acres', ascending=True).head(10)

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.barh(top_10['hood'], top_10['acres'], color='blue')  # Horizontal bar for readability
plt.xlabel('Acres')
plt.ylabel('Neighborhood')
plt.title('Top 10 Neighborhoods by Acres (Ascending Order)')
plt.tight_layout()

# show the plot
plt.show()
```
